```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
print('hi')
driver = webdriver.Chrome()
driver.get('https://m.imdb.com/title/tt0468569/?ref_=chttp_t_3')
y = 1000
for timer in range(0, 6):
    driver.execute_script("window.scrollTo(0, " + str(y) + ")")
    y += 1000
    time.sleep(2)
genre = driver.find_element(By.XPATH,"//li[@data-testid='storyline-genres']//div[@class='ipc-metadata-list-item__content-container']").text
                            
print(genre)
```
