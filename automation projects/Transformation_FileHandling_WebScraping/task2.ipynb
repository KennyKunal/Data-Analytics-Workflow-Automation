```python
import json
import csv
```

```python
with open("C:\\Users\\Kunal.Keni\\Desktop\\test.json") as f:
    data = json.load(f)
result = []
for i in data.values():
    l = []
    for j in i.values():
        # print(j)
        l.append(j)
    result.append(l)
```

```python
with open('test.csv','w',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(data.keys())
    for row in zip(*result):
        writer.writerow(row)
```
