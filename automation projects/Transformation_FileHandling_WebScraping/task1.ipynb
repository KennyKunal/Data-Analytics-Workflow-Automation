## Matrix addition and subtraction without using numpy array

```python
class Matrix():
    def __init__(self, matrix):
        self.matrix = matrix
        self.first_index = [0,1]
        self.second_index =[0,1]

    def __add__(self, other):
        if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]) and len(self.matrix[1]) == len(other.matrix[1]) :
            result = []
            for i in self.first_index:
                row = []
                for j in self.second_index:
                    val = self.matrix[i][j] + other.matrix[i][j]
                    row.append(val)
                result.append(row)
            return result
        else:
            return 'matrix should have same size'

    def __sub__(self, other):
        if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]) and len(self.matrix[1]) == len(other.matrix[1]):
            result = []
            for i in self.first_index:
                row = []
                for j in self.second_index:
                    val = self.matrix[i][j] - other.matrix[i][j]
                    row.append(val)
                result.append(row)
            return result
        else:
            return 'matrix should have same size'
```

```python
a = Matrix([[1, 2], [3, 4]])
b = Matrix([[4, 5], [6, 7]])
```

```python
a+b
```

```python
a-b
```
