```python


# Determine the maximum number of sellers
max_sellers = max(number_of_sellers)

# Create headers for CSV file
headers = ['Name', 'Final Price', 'Rating', 'Sold By']
for i in range(1, max_sellers + 1):
    headers.extend([f'Seller {i}', f'Price {i}', f'Type {i}'])

# Create a list of rows for the CSV file
rows = []
for i in range(len(name_list)):
    num_sellers = number_of_sellers[i]
    
    # Initialize row with item-specific data
    row = [name_list[i], final_price_list[i], rating_list[i], sold_by_list[i]]
    
    # Add data for each seller, price, and type
    for j in range(max_sellers):
        if j < num_sellers:
            # If seller, price, and type exist for the item, add them
            seller = sellers_list[i] if j < len(sellers_list[i]) else 'NaN'
            price = price_list[i] if j < len(price_list[i]) else 'NaN'
            type_ = type_list[i] if j < len(type_list[i]) else 'NaN'
        else:
            # If no seller, price, and type for the item, fill with 'NaN'
            seller, price, type_ = 'NaN', 'NaN', 'NaN'
        
        row.extend([seller, price, type_])
    
    # Add the row to the list of rows
    rows.append(row)

# Write the data to a CSV file
with open('amazon_data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(rows)
```
